# PACE: Construct

```python
# Dropping column 'label' because it is less than 5% og the data and no evidence of a non-random cause\
df = df.dropna(subset=['label'])
```

```python
'''Imputing the outliers by using the 95th percentile (the value below which 95% of the data falls). Essentially, replacing values that exceed the 95th percentile with the 95th percentile value -> bringing 
the outliers closer to the bulk of the data while still keeping some of the outliers influence.'''

# Imputing the outliers
for column in ['sessions', 'drives', 'total_sessions', 'total_navigations_fav1',
               'total_navigations_fav2', 'driven_km_drives', 'duration_minutes_drives']:
    threshold = df[column].quantile(.95)
    df.loc[df[column] > threshold, column] = threshold


```

```python
df.describe()
```

```python
'''Now changonf the data type of the 'label' column to the binary <- to use logistic regression model'''
'''0 are retained users and 1 are churned users'''

# Creting new binary column 'label2' 
df['label2'] = np.where(df['label']=='churned', 1, 0)
df[['label', 'label2']].tail()
```

```python
''' Determinig whether assumptions for logidtic regression have been met. '''
''' Checking the collinearity '''

# First, generating a correlation matrix 
numeric_columns = df.select_dtypes(include=['int64', 'float64'])
numeric_columns.corr(method='pearson')
```

```python
# Plotting correlation heatmap
plt.figure(figsize=(15,10))
sns.heatmap(numeric_columns.corr(method='pearson'), vmin=-1, vmax=1, annot=True, cmap='coolwarm')
plt.title('Correlation heatmap indicates many low correlated variables',
          fontsize=18)
plt.show();
```

Arbitraty threshold of 0.7 of the absolute value (other industrie might use .6, or .8, etc).

There are variables that are multicollinear with each other.
    sessions and drives: 1.0 
    driving_days and activity_days: 0.95

```python
# Creating a new binary column 'device2' using OneHotEncoder to change the varible from categorical
df['device2'] = np.where(df['device'] == 'Android', 0, 1)
df[['device', 'device2']].tail()
```

```python
''' Creating a model to assign predictor variables and targert'''
''' My model to predict the target 'label2,' I'll drop and then assign the results to X:

        label (this is the target)
        label2 (this is the target)
        device (this is the non-binary-encoded categorical variable)
        sessions (this had high multicollinearity)
        driving_days (this had high multicollinearity)
        '''

# Isolating predictor variables 
X = df.drop(columns = ['label', 'label2', 'device', 'device2', 'sessions', 'driving_days'])
```

```python
# Isolating the dependent target variable -> assign to y.
y = df['label2']
```

```python
# Performing the train-test split (the target class is imbaleced 82% retained vs 18% churned) to make sure that the split is not over- or under-represented the frequency of the minority class
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state= 42)

X_train.head()
```

```python
# Instantiating a logic regression model & adding pelanty= 'none' because my predictors are unscaled 
model = LogisticRegression(penalty=None, max_iter=400)
model.fit(X_train, y_train)
```

```python
pd.Series(model.coef_[0], index=X.columns)
```

```python
# Getting the intercept of the model by calling the model's intercept_
model.intercept_
```

```python
# Verifying the linear relationship  between X and the estimated log odds by making a regplot
# Getting predicted probablities of the trainnig data
training_probabilities = model.predict_proba(X_train)
training_probabilities 
```

```python
# Copying the 'X_train' data and assigning to 'logit_data'
logit_data = X_train.copy()

# Creating a new column called 'logit' 
logit_data['logit'] = [np.log(prob[1] / prob[0]) for prob in training_probabilities]
```

```python
# Plotting the reglot of the 'activity_days' log-odds
sns.regplot(x = 'activity_days', y = 'logit', data = logit_data, scatter_kws={'s':2, 'alpha':.5})
plt.title('log-odds: activity_days');
```

# PACE: Execute


```python
# Predictions on X_test
y_preds = model.predict(X_test)
```

```python
# Using score meethod to find the accuracy of the model
model.score(X_test, y_test)
```

```python
# Dispaying the confusion metrix
cm = confusion_matrix(y_test, y_preds)
```

```python
disp = ConfusionMatrixDisplay(confusion_matrix=cm, 
                            display_labels = ['retained', 'churned'],
                            )
disp.plot();
```

```python
# Calculating precision
precision = cm[1, 1] / (cm[0, 1] + cm[1, 1])
precision
```

```python
# Calculating recall
recall = cm[1, 1] / (cm[1, 0] + cm[1, 1])
recall
```

```python
# Creatign a classification report
target_labels = ['retained', 'churned']
print(classification_report(y_test, y_preds, target_names=target_labels))
```

```python
# Creating a list of (column_name, coefficient) tuples
feature_importance = list(zip(X_train.columns, model.coef_[0]))

# Sorting the list by coefficient value
feature_importance = sorted(feature_importance, key=lambda x: x[1], reverse=True)
feature_importance
```

```python
# Plotting the feature importances
import seaborn as sns
sns.barplot(x=[x[1] for x in feature_importance],
            y=[x[0] for x in feature_importance],
            orient='h')
plt.title('Feature importance');
```
